# tidy data

```python
df = df_raw.melt(id_vars=['국가(대륙)별', '상품군별', '판매유형별'], var_name='기간', value_name = '백만원')
df.shape
```

```python
df.head()
```

```python
df.info()
```

# 기간에서 연도, 분기 분리하기

```python
int(('2019 4/4').split()[0])
```

```python
df['연도'] = df['기간'].map(lambda x: int(x.split()[0]))
df.head()
```

```python
'2019 4/4'.split()[1]
```

```python
'2019 4/4'.split()[1].split('/')[0]
```

```python
df['분기'] = df['기간'].map(lambda x: int(x.split()[1].split('/')[0]))
df.head()
```

# 금액을 수치데이터로 표현하기 위해 데이터 타입 변경

```python
df['백만원']
```

```python
df['백만원']= df['백만원'].replace('-', pd.np.nan).astype(float)
df['백만원']
# 데이터 타입을 변경하기 위해 결측치를 nan으로 바꾼다
```

```python
df.head()
```

# 필요없는 데이터 제거하기

```python
df.info()
```

```python
# 합계 데이터 삭제
# ()&() 해야지 두가지 동시에 보기 가능, 
df = df[(df['국가(대륙)별']!='합계') & (df['상품군별']!='합계')].copy()
df.info()
# 데이터를 제거했기 때문에 메모리가 줄어듬
```

# 시각화 

```python
# 판매유형별 데이터는 일부 기간에는 계만 존재하기 때문에
# 판매유형별 == '계' 데이터만 가져와서 봐야지
# 평균 값을 구하는 그래프에서 올바른 값을 표현할 수 있다
df_total = df[df['판매유형별']=='계'].copy()
df_total.head()
```

```python
# lineplot 로 그리기
sns.lineplot(data= df_total, x='연도', y='백만원')
```

```python
# 상품군별로 다른 색상 칠하기
sns.lineplot(data=df_total, x='연도', y='백만원', hue='상품군별')
# legend 값을 바깥쪽으로 하기
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# 위에 그린 그래프를 자세히 보기 위해 서브플롯으로 표시
sns.relplot(data=df_total, x='연도', y='백만원', hue='상품군별', kind='line', col='상품군별', col_wrap=4)
```

```python
# 화장품 값이 너무 커서 다른 값들을 제대로 알 수 없음
# isin 을 이용해 화장품만 제외하고 df_sub 변수에 담자
# ~ 하면 그것만 빼고라는 뜻임
df_sub = df_total[~df_total['상품군별'].isin(['화장품'])].copy()
```

```python
sns.relplot(data=df_sub, x='연도', y='백만원', hue='상품군별', col='상품군별', col_wrap=4, kind='line')
```
